```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

from const import DATA_DIR

pio.templates.default = "plotly_dark"

# 2. Testing the Earth
tests_path = DATA_DIR / 'nuclear-weapons' / 'number-of-nuclear-weapons-tests' / 'number-of-nuclear-weapons-tests.csv'
df_tests = pd.read_csv(tests_path)

# Sort entities by total tests for legend order
total_tests = df_tests.groupby('Entity')['Number of nuclear weapons tests'].sum()
sorted_entities = total_tests.sort_values(ascending=False).index.tolist()

fig2 = px.bar(df_tests, 
              x='Year', 
              y='Number of nuclear weapons tests', 
              height=500,
              width=1000,
              color='Entity',
              labels={'Entity': "Country"},
              title='<b>Echoes of Detonations: Nuclear Weapons Tests</b><br><sub>Peak testing occurred during the height of Cold War tensions</sub>',
              category_orders={'Entity': sorted_entities},
              )

fig2.update_layout(barmode='stack', xaxis_title=None, hovermode='x unified')
fig2.update_traces(hovertemplate='%{y}')
fig2.show()
```

```python
from const import VISUALIZATIONS_DIR

fig2.write_html(VISUALIZATIONS_DIR / "echoes-of-detonations.html")
```
